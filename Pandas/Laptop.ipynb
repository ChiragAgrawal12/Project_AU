```python
def clean_column(ele):
    ele=ele.strip()                      # Remove leading and trailing whitespace
    ele=ele.lower()                      # Convert to lowercase
    ele=ele.replace(' ','_')             # Replace spaces with underscores
    ele=ele.replace('(','')              # Remove left parentheses
    ele=ele.replace(')','')              # Remove right parentheses
    ele=ele.replace('operating_system','os')  # Abbreviate 'operating_system' to 'os'
    return ele
```

```python
df.columns=[clean_column(col) for col in df.columns]   # Clean all column names
df.columns
```

```python
df['storage'].unique()
```

```python
import re
def to_gb(text):
    parts = re.findall(r'(\d+(?:\.\d+)?)\s*(GB|TB)', text)
    total = 0
    for num, unit in parts:
        num = int(num)
        if unit == 'TB':
            num *= 1024
        total += num
    return total

df['storage'] = df['storage'].apply(to_gb)
df['storage']
```
