```python
import pandas as pd
import numpy as np
```

```python
p1=pd.Series([100,200,300],index=['a','b','c'])       # Creating a Pandas Series with custom index
p1
```

```python
p1.index=['l','m','n']
p1
```

```python
p1[0]                     # Accessing element by position as index which has been changed
```

```python
p1['l']
```

```python
p1.sum()                    # Sum of all elements in the Series
```

```python
p1.min()                    # Minimum element in the Series
```

```python
p1.max()                    # Maximum element in the Series
```

```python
type(p1)                    # Checking the type of p1
```
