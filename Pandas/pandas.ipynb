```python
df1=pd.DataFrame({'Name':['Aaditya','Muskan','Chirag','Vaibhav'],'Marks':[92,98,90,84],'Subjects':['Maths','Physics','Chemistry','Biology']})                             # Creating a Pandas DataFrame
df1
```

```python
df1.index                        # Checking the index of the DataFrame
```

```python
df1.columns                      # Checking the columns of the DataFrame
```

```python
df1['Marks']                     # Accessing a column in the DataFrame
```

```python
df1['Name']=['A','M','C','V']   # Modifying a column in the DataFrame
df1
```

```python
df1.iloc[0:2,0:2]                   # iloc( row start index: row stop index : step, column start index: column stop index : step) Accessing rows by integer-location based indexing 
```

```python
df1.loc[0:2,'Name']                  # loc( row labels , column labels) Accessing rows by label-based indexing
```

```python
# Create a sample DataFrame
data = {
    'Name': ['Amit', 'Priya', 'Rahul', 'Sneha', 'Karan'],
    'Age': [23, 25, 22, 24, 26],
    'City': ['Delhi', 'Mumbai', 'Pune', 'Bangalore', 'Chennai'],
    'Score': [85, 90, 78, 88, 92]
}

df2 = pd.DataFrame(data, index=['s1', 's2', 's3', 's4', 's5'])
df2
```

```python
df2.iloc[0::2,0::2]               
```

```python
df2.iloc[0:2,0]
```

```python
df2.iloc[0:,[0,2]]               # Accessing specific rows and columns using iloc integer location based indexing
```

```python
df2.loc['s1':'s3','Name':'City']        # Accessing specific rows and columns using loc location based indexing
```

```python
df2.loc['s1':'s3',['Name','City']]
```

```python
df2.columns
```

```python
df2['Age']
```

```python
cond=df2['Age']==25
cond
```

```python
X=df2.loc[cond]
X
```

```python
X.loc[:,'Age']=22
X
```

```python
df2.loc[df2['Age']==25,'Age']=22
df2
```

```python
df2.loc[df2['City']=='Mumbai','City']='Kolkata'
df2
```

```python
df2.loc[(df2['Age']>20) & (df2['Age']<=25),'Age']=22
df2
```
