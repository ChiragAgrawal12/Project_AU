## Line Plot

#### Parameters in plot function

* ls - line style 
* lw - line width
* color - line color
* label - legend label
* alpha - grid transparency
* marker - marker type
* markersize - marker size
* markerfacecolor - marker face color
* markeredgecolor - marker edge color
* fontsize - font size
* fontstyle - font style
* axis - axis range
* title - plot title
* xlabel - x axis label
* ylabel - y axis label
* legend - legend location
* grid - grid visibility
* figsize - figure size
* dpi - figure resolution
* facecolor - figure background color


```python
plt.plot(China['Date_reported'],China['Cumulative_cases'],ls="--",color="red",label="China")
plt.plot(India['Date_reported'],India['Cumulative_cases'],label="India")
plt.xlabel('Date')
plt.ylabel('Cummulative_cases')
plt.title('China Covid Cases')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
def Data_fetcher(*args):
    for i in args:
        kwargs=df[df['Country']==i]
        plt.plot(kwargs['Date_reported'],kwargs['Cumulative_cases'])
    plt.xlabel('Date')
    plt.ylabel('Cummulative_cases')
    plt.title('Covid Cases')
    plt.xticks(rotation=90)
    plt.legend([i for i in args])
    plt.show()
```

```python
Data_fetcher('India','China','France')
```

```python
plt.Figure(figsize=(9,9))
plt.subplots(1,3)
# countries=["China",'India','Pakistan']
color=['Red','Blue','Green']
def plot_btw_countries(countries):
    for i in range(len(countries)):
        plt.subplot(1,3,i+1)
        plt.title(f"{countries[i]}-cases")
        country_df = df[df['Country'] == countries[i]]
        plt.plot(country_df['Date_reported'], country_df['Cumulative_cases'],color=color[i], label=countries[i])
        plt.xlabel("Date", fontsize=12)
        plt.ylabel("Cumulative Cases", fontsize=12)
        plt.xticks(rotation=90)
        plt.grid(True, linestyle=":", alpha=0.5)
        plt.legend()
        plt.tight_layout()
    plt.show()
plot_btw_countries(["China",'India','Pakistan'])
```

```python
import matplotlib.patches as mpatches

red_patch = mpatches.Patch(color='red', label='Red')
blue_patch = mpatches.Patch(color='blue', label='Blue')
green_patch = mpatches.Patch(color='green', label='Green')
pink_patch = mpatches.Patch(color="#fb2be9", label='Pink')

Set1=['blue','green','red','#fb2be9']
data=day[['season','cnt']].groupby(['season'])['cnt'].sum()
plt.pie(data,autopct='%1.1f%%',explode=[0.1,0.1,0.1,0.1],labels=["Winter","Spring","Summer","Autumn"],colors=Set1)
plt.legend(handles=[blue_patch, green_patch, red_patch, pink_patch])
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# legend patches
blue_patch = mpatches.Patch(color='blue', label='Winter')
green_patch = mpatches.Patch(color='green', label='Spring')
red_patch = mpatches.Patch(color='red', label='Summer')
pink_patch = mpatches.Patch(color='#fb2be9', label='Autumn')

# color map for slices (matches legend order!)
Set1=['blue','green','red','#fb2be9']

# aggregated data
data = day[['season','cnt']].groupby('season')['cnt'].sum()

# draw pie chart
plt.pie(
    data,
    autopct='%1.1f%%',
    explode=[0.1,0.1,0.1,0.1],
    labels=["Winter","Spring","Summer","Autumn"],
    colors=Set1
)

# add matching legend
plt.legend(handles=[blue_patch, green_patch, red_patch, pink_patch])

plt.show()
```
