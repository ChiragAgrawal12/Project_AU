## Line Plot

#### Parameters in plot function

* ls - line style 
* lw - line width
* color - line color
* label - legend label
* alpha - grid transparency
* marker - marker type
* markersize - marker size
* markerfacecolor - marker face color
* markeredgecolor - marker edge color
* fontsize - font size
* fontstyle - font style
* axis - axis range
* title - plot title
* xlabel - x axis label
* ylabel - y axis label
* legend - legend location
* grid - grid visibility
* figsize - figure size
* dpi - figure resolution
* facecolor - figure background color


```python
plt.plot(China['Date_reported'],China['Cumulative_cases'],ls="--",color="red",label="China")
plt.plot(India['Date_reported'],India['Cumulative_cases'],label="India")
plt.xlabel('Date')
plt.ylabel('Cummulative_cases')
plt.title('China Covid Cases')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
def Data_fetcher(*args):
    for i in args:
        kwargs=df[df['Country']==i]
        plt.plot(kwargs['Date_reported'],kwargs['Cumulative_cases'])
    plt.xlabel('Date')
    plt.ylabel('Cummulative_cases')
    plt.title('Covid Cases')
    plt.xticks(rotation=90)
    plt.legend([i for i in args])
    plt.show()
```

```python
Data_fetcher('India','China','France')
```

```python
day=pd.read_csv('day.csv')
day['dteday']=pd.to_datetime(day['dteday'])
```

```python
day.head()
```

```python
day.info()
```

```python
day.groupby('season')[['temp','hum','windspeed']].mean()
```

```python
plt.scatter(day['dteday'],day['cnt'],c=day['season'],cmap='jet')
plt.colorbar()
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

### <b><u> METHOD 1

```python

for i in day['season'].unique():
    plt.scatter(day[day['season']==i]['dteday'],day[day['season']==i]['cnt'],label=i)
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(['Winter','Spring','Summer','Autumn'])
plt.show()
```

### <b><u> METHOD 2

```python
seasons=day['season'].unique()
weather=['Winter','Spring','Summer','Autumn']
for i in range(len(seasons)):
    new_df=day[day['season']==seasons[i]]
    plt.scatter(new_df['dteday'],new_df['cnt'],label=weather[i])
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
