#### <b><u>Boolean Masking

```python
n1=np.arange(2,20)
n1
```

```python
condition=n1>7
condition
```

```python
n1[condition]
```

```python
n1[n1>10]                    # In boolean masking we can directly use the condition inside the indexing brackets to filter elements
```

```python
n1[(n1>10) & (n1<16)]
```

```python
n1[(n1<10) | (n1>16)]
```

```python
fruits=np.array(["apple","mango","banana","chilli"])
price=np.array([20,30,4,6])
```

```python
fruits[price<20]      # In this case boolean masking is applied on one array (price) to filter elements from another array (fruits) and will only work if length of both arrays is same
```

```python
n2=np.arange(10,16)
n2
```

```python
np.where(n2%2==0,"even","odd")     # np.where(condition, value if true, value if false) it is similar to if else statement
```

#### <b><u> Broadcasting in Numpy

```python
n23=np.arange(1,5).reshape(2,2)      # broadcasting helps to perform operations on arrays of different shapes by automatically expanding the smaller array to match the shape of the larger array
n24=np.arange(1,3)
```

```python
n23
```

```python
n24
```

```python
n23+n24                              
```

```python
A=np.asarray([[30,-1],[50,-1]])
A
```

```python
InvA=np.linalg.inv(A)            # np.linalg.inv() computes the inverse of a square matrix
InvA
```

```python
B=np.asarray([-1000,-100])
B
```

```python
X=np.dot(InvA,B)                    # np.dot() performs matrix multiplication as dor product of two arrays
X
```
