### Q3

```python
def intostr(i):                      # It has time complexity O(log(n))
  digit="0123456789"
  if i==0:
    return "0"
  result=" "
  while i>0:
    result=digit[i%10]+result
    i=i//10
  return result
```

### Q8

```python
def factorial(n):                    # It has time complexity O(n)
  if n==1:
    return 1
  else:
    return n*factorial(n-1)
```

### Q9

```python
def fib(n):                          # It has time complexity O(2^n)
  if n==1 or n==0:
    return 1
  else:
    return fib(n-1)+fib(n-2)
```

### Q10

```python
def bubble_sort(arr):                            # It has time complexity O(n^2)
    for i in range(len(arr)-1):
        flag=0
        for j in range(len(arr)-1-i):
            if arr[j+1]<arr[j]:
                arr[j+1],arr[j]=arr[j],arr[j+1]
                flag=1
        if flag==0:
            break
    print(arr)

```

```python
def binary_search(arr,item):              # It has time complexity O(log(n))
  low=0
  high=len(arr)-1
  while low<=high:
    mid=(low+high)//2
    if arr[mid]==item:
      return mid
    elif arr[mid]>item:
      high=mid-1
    elif arr[mid]<item:
      low=mid+1
  return "not found"
print(binary_search([1,2,3,4,5,6,78],100))
```
